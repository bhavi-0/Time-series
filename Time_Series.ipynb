### Implementing ML Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import math
from lightgbm import LGBMRegressor
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
df.columns = df.columns.str.replace('Unnamed: 0', 'Group')
df_test.columns = df_test.columns.str.replace('Unnamed: 0', 'Group')
etr1 = GradientBoostingRegressor(random_state=30, verbose=1)
```

```python
etr2 = RandomForestRegressor(n_estimators = 100, random_state = 0)
```

```python
etr= LGBMRegressor()
```

```python
predicted = {'Para-9': [], 'Para-10': [], 'Para-11': [], 'Para-12': [], 'Para-13': []}
names = ['Para-9', 'Para-10', 'Para-11', 'Para-12', 'Para-13']
```

```python
rmse = []
N = 0
index = 0
for j in [9,10, 11, 12, 13]:
  ans = 0
  count = 0
  for i in range(0, 1000, 10):
    xtrain = df.iloc[i:i+9, 1:10]
    ytrain = df.iloc[i:i+9,j ]
    xtest = df.iloc[i+9, 1:10].to_numpy()
    ytest = df.iloc[i+9, j]
    etr.fit(xtrain, ytrain)
    ypred = etr.predict(xtest.reshape(1,-1))[0]
    predicted[names[index]].append(ypred)
    ans += (ytest - ypred)**2
    count += 1
    N += 1
  rmse.append(ans)
  index += 1
```

```python
df_prediction = pd.DataFrame.from_dict(predicted)
df_prediction
```

```python
for i in rmse:
  print("RMSE = ",(i/count)**0.5)
```

```python
print("Total RMSE = ",(sum(rmse)/N)**0.5)
```

### Plots

```python
for i in range(5):
    plt.figure(figsize=(6,6))
    true_value = df_prediction.iloc[:, i]
    predicted_value = df_test.iloc[:, i + 10]
    plt.scatter(true_value, predicted_value, c='orange')

    p1 = max(max(predicted_value), max(true_value))
    p2 = min(min(predicted_value), min(true_value))
    plt.plot([p1, p2], [p1, p2], 'b-')
    plt.title(names[i])
    plt.xlabel('True Values', fontsize=10)
    plt.ylabel('Predictions', fontsize=10)
    plt.axis('equal')
    plt.show()
```
